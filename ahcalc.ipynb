```python
def tokenize(input_string):
    
    #Validate for illegal characters
    if re.search(r'''[^0-9\.\+\-\*/\(\)\[\]\!\^\s]''', input_string, re.VERBOSE):
        raise ValueError, 'The only allowable characters are 0-9, +, -, *, /, !, ^, parentheses, and square brackets.'
        return

    #Validate for unbalanced brackets
    if (input_string.count('(') != input_string.count(')') or \
            input_string.count('[') != input_string.count(']')):
    
        raise ValueError, "Unbalanced brackets or parentheses."

    
    # Don't want whitespace
    regularized_string = re.sub(r'\s', '', input_string)

    # Make everything use normal parentheses
    regularized_string = regularized_string.translate(string.maketrans('[]', '()'))
    
    # Easy way to convert negative numbers into a compatible form for parsing.
    # Convert hyphen-number at the start of the string, start of brackets, or after another operator strings like this:
    # -5+6 becomes (0-5)+6
    # (-5+6) becomes ((0-5)+6)
    # 5*-6 becomes 5*(0-6)
    # Unclear why, but the substitution doesn't work as expected unless the regex is pre-compiled.
    negative_number_regex = re.compile(r'''(^-[\d\.]+|    (?# Negative number at start of the line)
                                           (?<=[-+*/(])-[\d\.]+) (?# Negative number after other operator or start of parens)
                                           ''', re.VERBOSE)
    regularized_string = re.sub(negative_number_regex, r'(0\1)', regularized_string, re.VERBOSE)
    
    # Similar way to convert negative parentheses into a compatible form for parsing.
    # Convert '-(' in places where it signifies a negative:
    # -(1+2) becomes (0-1)*(1+2)
    
    regularized_string = re.sub(r'(?<=[-+*/(])-\(', r'(0-1)*(', regularized_string, re.VERBOSE) # after operator
    regularized_string = re.sub(           r'^-\(', r'(0-1)*(', regularized_string, re.VERBOSE) # at start of line
    
    # Regularize multiplication in the form 2(3+4), (2+3)4, or (2+3)(4+5)
    regularized_string = re.sub(r'''(\d)(\()''', r'''\1*\2''', regularized_string)
    regularized_string = re.sub(r'''(\))(\d)''', r'''\1*\2''', regularized_string)
    regularized_string = re.sub(r'''(\))(\()''', r'''\1*\2''', regularized_string)
    
    print(regularized_string)
    
    tokenized_input = re.split(r'''([^\d\.])''', regularized_string, re.VERBOSE)
    # Discard null strings which result from matching split patterns at the end of the string (close brackets and !.)
    while '' in tokenized_input:
        tokenized_input.remove('')
    
    tokenized_input_cast = []
    for token in tokenized_input:
        try:
            token_cast = float(token) 
        except ValueError:
            token_cast = token
        
        tokenized_input_cast.append(token_cast)
    
    
    
    return tokenized_input_cast
```

```python
def buildtree(tokens):
    print "Tokens: ", tokens
    math_funcs = {'^': pow,
                  '+': operator.add,
                  '-': operator.sub,
                  '!': math.factorial,
                  '*': operator.mul,
                  '/': operator.div}
    tree = tokens[:]
    
    # Recursive case: there is at least one level of nested parentheses to process.
    while '(' in tree:
#        import pdb; pdb.set_trace()
        paren_start = tree.index('(')
        paren_contents = []
        depth = 0
        
        for i, token in enumerate(tree[paren_start:]):
            if token == '(':
                depth += 1
            elif token == ')': 
                depth -= 1
            if depth == 0:
                paren_len = i
                break
        # The for loop never hit a break, which means we didn't find a matching close parenthesis.
        else:
            raise ValueError, "Unmatched parentheses."
        
        paren_end = paren_start + paren_len
        
        # recursion magic: support arbitrarily deeply nested sets of parentheses
        tree[paren_start:paren_end+1] = [buildtree(tree[paren_start+1:paren_end])]
    
    # Base case: there are no parentheses in this list, and we can group the other operators and go home.
    
    while '!' in tree:
        fact_idx = tree.index('!')
        if type(tree[fact_idx-1]) not in [float, list] or fact_idx==0:
            raise ValueError, "Factorial without an appropriate group or number preceding it."
        else:
            tree[fact_idx-1:fact_idx+1] = [[math_funcs['!'], tree[fact_idx-1]]]
            
    while '^' in tree:
        exp_idx = tree.index('^')
        if (type(tree[exp_idx-1]) not in [float, list] or
               type(tree[exp_idx+1]) not in [float, list] or
               exp_idx in [0, len(tree)-1]):
            raise ValueError, "Exponent without appropriate groups or numbers before and after it."
        tree[exp_idx-1:exp_idx+2] = [[math_funcs['^'], tree[exp_idx-1], tree[exp_idx+1]]]
    
    while ('*' in tree) or ('/' in tree):
        
        # Which operator did we catch? If multiple, select the first one.
        if '*' in tree and '/' in tree:
            muldiv_idx = min(tree.index('*'), tree.index('/'))
        elif '/' in tree:
            muldiv_idx = tree.index('/')
        elif '*' in tree:
            muldiv_idx = tree.index('*')
        else:
            raise ValueError
        
        muldiv_char = tree[muldiv_idx]
        if (type(tree[muldiv_idx-1]) not in [float, list] or
               type(tree[muldiv_idx+1]) not in [float, list] or
               muldiv_idx in [0, len(tree)-1] ):
            raise ValueError, "* or / without appropriate groups or numbers before and after it."
        tree[muldiv_idx-1:muldiv_idx+2] = [[math_funcs[muldiv_char], tree[muldiv_idx-1], tree[muldiv_idx+1]]]
    
    while ('+' in tree) or ('-' in tree):
        
        # Which operator did we catch? If multiple, select the first one.
        if '+' in tree and '-' in tree:
            addsub_idx = min(tree.index('+'), tree.index('-'))
        elif '+' in tree:
            addsub_idx = tree.index('+')
        elif '-' in tree:
            addsub_idx = tree.index('-')
        else:
            raise ValueError
        
        addsub_char = tree[addsub_idx]
        if (type(tree[addsub_idx-1]) not in [float, list] or
               type(tree[addsub_idx+1]) not in [float, list] or
               addsub_idx in [0, len(tree)-1]):
            raise ValueError, "+ or - without appropriate groups or numbers before and after it."
        tree[addsub_idx-1:addsub_idx+2] = [[math_funcs[addsub_char], tree[addsub_idx-1], tree[addsub_idx+1]]]
    

    # We have consumed all the operators in order of precedence, and at the end there's
    # a nicely-arranged tree we can pass to evaltree.
    print tree
    return tree
```

```python
def evaltree(expr):
#     recursive case: argument is a tree.
#     base case: argument is a float.
    if type(expr) == float:
        return expr
    elif len(expr) == 1 and type(expr[0]) == float:
        return expr[0]
    elif len(expr) == 1 and type(expr[0]) == list:
        return evaltree(expr[0])
    elif len(expr) == 2:
        return expr[0](evaltree(expr[1])) # Factorials are the only unary operators being supported now.
    elif len(expr) == 3:
#        import pdb; pdb.set_trace()
        return expr[0](evaltree(expr[1]), evaltree(expr[2]))
    else:
        print(expr)
        print("Length: ", len(expr))
#        import pdb; pdb.set_trace()
        raise ValueError, "evaltree received a list with an invalid length."
```
